```python
data = pd.read_csv("/Users/eren/Downloads/data.csv")
```

```python
data.drop(["id","Unnamed: 32"], axis = 1, inplace= True)
```

```python
M = data[data.diagnosis == "M"]
B = data[data.diagnosis == "B"]
```

```python
plt.scatter(M.radius_mean,M.texture_mean,color="red",label="tehlikeli",alpha=0.3)
plt.scatter(B.radius_mean,B.texture_mean,color="green",label="zararrsÄ±z",alpha=0.3)
plt.xlabel("radius_mean(tÃ¼mÃ¶r yarÄ±Ã§apÄ±)")
plt.ylabel("texture_mean(tÃ¼mÃ¶r dokusu)")
plt.show()
```

```python
data.diagnosis = [1 if each == "M" else 0 for each in data.diagnosis]
x_data = data.iloc[:,1:3].values
y = data.diagnosis.values
```

```python
from sklearn.model_selection import train_test_split

x_train, x_test, y_train, y_test = train_test_split(x_data,y,test_size=0.3,random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
st_x=StandardScaler()
```

```python
x_train = st_x.fit_transform(x_train)
```

```python
x_test = st_x.transform(x_test)
```

```python
from sklearn.tree import DecisionTreeClassifier
tree_classification = DecisionTreeClassifier(random_state=1,criterion='entropy')
tree_classification.fit(x_train,y_train)
```

```python
y_head = tree_classification.predict(x_test)
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy = accuracy_score(y_test,y_head)
accuracy
```

```python
from sklearn.metrics import confusion_matrix
cm = confusion_matrix(y_test,y_head)
```

```python
f,ax = plt.subplots(figsize=(5,5))
sns.heatmap(cm,annot=True,fmt=".0f",linewidths=0.5,linecolor="red", ax=ax)
plt.xlabel("y_pred")
plt.ylabel("y_test")
plt.show()
```
